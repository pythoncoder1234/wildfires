```python
import opendatasets as od
import pandas as pd

download = False

if download:
    od.download("https://www.kaggle.com/datasets/capcloudcoder/us-wildfire-data-plus-other-attributes/data")
    df = pd.read_csv("us-wildfire-data-plus-other-attributes/FW_Veg_Rem_Combined.csv")
else:
    df = pd.read_csv("us-wildfire-data.csv")
    
df = df[["fire_size", "fire_size_class", "latitude", "longitude", "Vegetation", "fire_mag", *df.columns[10:]]]
```

```python
bins = [0, 1, 2, 6, 99999999]
df['fire_size_class'] = pd.cut(df['fire_size'], bins=bins, labels=[i for i in range(len(bins)-1)])
```

```python
from matplotlib import pyplot as plt

plt.bar(df['fire_size_class'].value_counts().index, df['fire_size_class'].value_counts())
plt.show()
```

```python
from pprint import pprint

all_cols = df.columns.tolist()
assert len(all_cols) == len(set(all_cols))
keep = ['latitude', 'longitude', 'Vegetation', 'fire_size_class']
normalize = list(filter(lambda c: c not in keep, all_cols))
minmax = []
for col in normalize:
    if df[col].std() > 4:
        minmax.append(col)
pprint(normalize, compact=True)
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

df_normalized = df.copy()
for col in normalize:
    df_normalized[col] = StandardScaler().fit_transform(df[col].to_numpy().reshape(-1, 1))
# for col in minmax:
#     df[col] = MinMaxScaler().fit(df[col])

df_normalized.describe()
df = df_normalized.copy()
```

```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split

seed = 123
batch_size = 256

x = df.iloc[:,2:].to_numpy()
y = df['fire_size_class'].to_numpy()

train_x, test_x, train_y, test_y = train_test_split(x, y, random_state=seed, test_size=0.2)
train_y = train_y.reshape(1, -1)[0] 
test_y = test_y.reshape(1, -1)[0]

# feature_batches, label_batches = split_into_batches(train_x, train_y, batch_size)
# print(list(map(lambda x: x.shape, feature_batches)), list(map(lambda x: x.shape, label_batches)))

run_count = 5
n_tr = np.linspace(batch_size, len(train_x), run_count, dtype=int)
train_acc = []
test_acc = []

for i, val in enumerate(n_tr):
    train_x2, train_y2 = map(lambda x: x[:val], [train_x, train_y]) 
    
    print(f"\nRun {i+1}, using {(val/len(train_x) * 100):.0f}% of all data")
    print(train_x2.shape, train_y2.shape, test_x.shape, test_y.shape)
    
    # model = MLPClassifier(hidden_layer_sizes=[64,32,8], random_state=seed, activation='relu',
    #                       solver='sgd', batch_size=256, learning_rate_init=0.001)
    
    model = KNeighborsClassifier(n_neighbors=20)
    
    model.fit(train_x2, train_y2)
    train_acc.append(model.score(train_x2, train_y2))
    test_acc.append(model.score(test_x, test_y))

print("\nTraining Accuracy:", ", ".join(map(lambda ac: f"{round(ac * 100, 1)}%", train_acc)))
print("Testing Accuracy:", ", ".join(map(lambda ac: f"{str(round(ac * 100, 1))}%", test_acc)))
plt.plot(n_tr, train_acc, label="Training Accuracy")
plt.plot(n_tr, test_acc, label="Testing Accuracy")
plt.xlabel('Number of samples')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

sklearn_disp = ConfusionMatrixDisplay(confusion_matrix = confusion_matrix(test_y, model.predict(test_x)))
sklearn_disp.plot()
```
